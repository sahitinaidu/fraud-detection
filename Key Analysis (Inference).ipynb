```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df2=pd.DataFrame()
df1=pd.DataFrame()
df = pd.read_csv("transactions.csv")
#print(df)

dfcust = pd.read_csv("customer.csv")
#print(dfcust)
df3 = dfcust.Fraud*1
#print(df3)

n= dfcust[dfcust['Fraud'] == 0]
rows=n.shape[0]
m=  dfcust[dfcust['Fraud'] == 1]
rows1 = m.shape[0]
li = [rows,rows1]
lis = ['NOT fraud','fraud']
print(li)
print(lis)
```

```python
plt = sns.barplot(x=lis, y=li, data=dfcust)
print("The number of Fraud Transaction")
```

```python
import matplotlib.pyplot as plt

piedata = df.paymentMethodType.value_counts()
figureObject, axesObject = plt.subplots()
colors = ['red', 'blue', 'yellow', 'pink']
labels = ['Card', 'Bitcoin', 'Paypal','Apple Pay']
explode = (0.2, 0.2, 0.2, 0.2)
plt.pie(piedata, colors=colors, labels=labels,
explode=explode, autopct='%1.1f%%',radius = 1,
counterclock=False, shadow=True)
plt.show()
print("The Transaction details through various modes")



```
